```python
import numpy as np
import pandas as pd
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
# считываем данные
train = pd.read_csv('./data/digit/train.csv')
test  = pd.read_csv('./data/digit/test.csv')
target = train["label"]
train = train.drop("label",1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(train)
transform = pca.transform(train)
```

```python
df = pd.DataFrame(transform)
```

```python
df['tg']= target
df['size'] = 1
```

```python
df.columns = ['1','2','3','tg','size']
```

```python
import plotly.express as px # визуализируем эту красоту
fig = px.scatter_3d(df[:1000], x='1', y='2', z='3', color='tg', size = 'size', size_max =5)
fig.show()
```

```python
pca = PCA(n_components=60)
pca.fit(train)
transform = pca.transform(train)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(transform, target, test_size=0.3, random_state=42)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train) # тренируем нашу модель
y_pred = log_reg.predict(X_test)
metrics.accuracy_score(y_pred, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier 

clf_rf = RandomForestClassifier(n_estimators=150, max_depth=100, min_samples_leaf=1, max_features=0.22, n_jobs=-1)
clf_rf.fit(X_train, y_train)
pred_val2 = clf_rf.predict(X_test)
metrics.accuracy_score(pred_val2, y_test) 
```

лес решений конечно отличный результат показал
